```python
# [セル1] テスト対象の関数（SQLの構文チェック）
def validate_sql(sql_query: str) -> bool:
    """SQL構文に SELECT と FROM が含まれているかチェックする関数"""
    query_upper = sql_query.upper()
    return "SELECT" in query_upper and "FROM" in query_upper


# [セル2] pytest用のテスト関数（ここがCIで自動検証されます）
def test_sql_has_select_and_from():
    # 正常系：正しくSELECTとFROMがある場合
    valid_query = "SELECT user_id, name FROM users WHERE age >= 20;"
    assert validate_sql(valid_query) is True

    # 異常系：FROMが抜けている場合
    invalid_query_1 = "SELECT user_id, name;"
    assert validate_sql(invalid_query_1) is False

    # 異常系：SELECTが抜けている場合
    invalid_query_2 = "INSERT INTO users (name) VALUES ('Taro');"
    assert validate_sql(invalid_query_2) is False

    # 正常系：小文字で書かれていても検知できるか
    lowercase_query = "select id from items;"
    assert validate_sql(lowercase_query) is True
```
