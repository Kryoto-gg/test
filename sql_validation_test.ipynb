```python
# [セル1] テスト対象の関数（SQLの構文チェック）
def validate_sql(sql_query: str) -> bool:
    """SQL構文に SELECT と FROM が含まれているかチェックする関数"""
    return "select" in sql_query and "from" in sql_query
```

```python
import os


def test_uploaded_sql_file():
    sql_file_path = "test.sql"

    # 【重要】現在の実行環境のディレクトリに何があるかログに出す
    print("\n--- 現在のフォルダにあるファイル一覧 ---")
    print(os.listdir("."))

    # ファイルを開いて中身を読み込む
    with open(sql_file_path, "r", encoding="utf-8") as f:
        sql_content = f.read()

    # 【重要】実際に読み込んだSQLの中身をログに強制出力する
    print("\n--- 読み込んだSQLの中身 ---")
    print(f"[{sql_content}]")
    print("---------------------------\n")

    # 構文チェックを実行
    is_valid = validate_sql(sql_content)

    # 判定結果も出力
    print(f"バリデーション結果: {is_valid}")

    # ここで確実に落とす設定
    assert (
        is_valid is True
    ), f"SQL構文エラー: {sql_file_path} に SELECT または FROM が含まれていません！"
```
