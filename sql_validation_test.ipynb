```python
# [セル1] テスト対象の関数（SQLの構文チェック）
def validate_sql(sql_query: str) -> bool:
    """SQL構文に SELECT と FROM が含まれているかチェックする関数"""
    query_upper = sql_query.upper()
    return "SELECT" in query_upper and "FROM" in query_upper
```

```python
import os


def test_uploaded_sql_file():
    # 1. チェックしたいSQLファイルのパスを指定
    sql_file_path = "test.sql"

    # 2. ファイルが存在するか確認（ファイルがない場合はテストをスキップ、または失敗させる）
    assert os.path.exists(
        sql_file_path
    ), f"{sql_file_path} が見つかりません。ファイルをアップロードしてください。"

    # 3. ファイルの中身を読み込む
    with open(sql_file_path, "r", encoding="utf-8") as f:
        sql_content = f.read()

    # 4. 読み込んだSQLに対して、SELECTとFROMがあるか検証する
    # ※今回の「test_col1 ...」のファイルなら、ここで正しく False（構文エラー）と判定され、
    # 『テストが失敗（意図通り異常を検知）』するようになります。
    assert (
        validate_sql(sql_content) is True
    ), f"SQL構文エラー: {sql_file_path} に SELECT または FROM が含まれていません！"
```
